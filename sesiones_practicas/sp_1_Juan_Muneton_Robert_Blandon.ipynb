## <span style="color:Blue;"><strong>4. Visualizaciones exploratorias b√°sicas</strong></span>

<div style="background-color:#FCE4EC; padding:10px; border-radius:8px;">
<b>- Gr√°ficos de dispersi√≥n para relaciones entre variables.</b><br>
<b>- Breve interpretaci√≥n de lo que muestran las gr√°ficas.</b><br>
<b>- Histogramas o gr√°ficos de barras para variables num√©ricas y categ√≥ricas.</b><br>

```python
df['Gross'] = df['Gross'].astype(str).str.replace(',', '')
df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')
```

```python
# Histograma: Distribuci√≥n de ingresos
plt.figure(figsize=(8,5))
sns.histplot(df['Gross'], bins=30)
plt.title("Distribuci√≥n de Ingresos")
plt.show()
```

```python
# Scatterplot: IMDB_Rating vs Gross
plt.figure(figsize=(8,5))
sns.scatterplot(x='IMDB_Rating', y='Gross', data=df)
plt.title("Relaci√≥n entre Calificaci√≥n IMDB e Ingresos")
plt.show()
```

```python
# Boxplot: Top 10 Directores vs Gross
top_directors = df['Director'].value_counts().head(10).index
df_top_directors = df[df['Director'].isin(top_directors)]

plt.figure(figsize=(12,6))
sns.boxplot(x='Director', y='Gross', data=df_top_directors)
plt.xticks(rotation=45)
plt.title("Ingresos seg√∫n Director (Top 10)")
plt.show()
```

```python
# Boxplot: Top 10 Actores (Star1) vs Gross
top_actors = df['Star1'].value_counts().head(10).index
df_top_actors = df[df['Star1'].isin(top_actors)]

plt.figure(figsize=(12,6))
sns.boxplot(x='Star1', y='Gross', data=df_top_actors)
plt.xticks(rotation=45)
plt.title("Ingresos seg√∫n Actor Principal (Top 10)")
plt.show()
```

```python
# Gr√°fico de barras: Promedio de ingresos por Director
director_gross = df.groupby('Director')['Gross'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
director_gross.plot(kind='bar')
plt.title("Top 10 Directores por Ingreso Promedio")
plt.ylabel("Ingreso Promedio")
plt.show()
```
