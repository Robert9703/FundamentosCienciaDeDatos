### <span style="color:#2F749F;"><strong>Ejercicio 1: Cálculo de medidas de tendencia central</strong></span>

A partir de una muestra simulada de ingresos, calcula la **media, mediana y moda**, y luego a interpretarlas.

1. Usa el número base `528310` (debes reemplazarlo con los últimos 6 dígitos de tu documento de identificación).  
2. Con este número genera una semilla (`np.random.seed`) para que los resultados sean reproducibles.  
3. Los ingresos se simulan con una distribución normal:  
   - Media teórica: `(base % 10) + 2`  
   - Desviación estándar: `2`  
   - Tamaño de muestra: `30`  
4. Los valores negativos se reemplazan por cero (pues los ingresos no pueden ser negativos).  
5. Grafica tus resultados (ver imagen de ejemplo)

<p align="center">
    <img src="https://github.com/mariabda2/intro_data_2026/blob/main/sesiones_practicas/imgs/sp_2_e1.png?raw=1" alt="sp_2_e1" width="900"  height="400">
</p>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

base = 167910
np.random.seed(base)
media_teorica = (base % 10) + 2
desviacion = 2
n = 30
```

```python
#Generar ingresos simulados
ingresos = np.array([x for x in np.random.normal(media_teorica, desviacion, n*2) if x > 0][:n])

#Calcular medidas
media = np.mean(ingresos)
mediana = np.median(ingresos)
moda = stats.mode(ingresos, keepdims=True)[0][0]

print("Media:", round(media,2))
print("Mediana:", round(mediana,2))
print("Moda:", round(moda,2))
```

```python
#Gráfica
plt.figure(figsize=(10,5))

sns.histplot(ingresos, bins=10, kde=True, color='skyblue')

plt.axvline(media, color='red', linestyle='--', label=f"Media = {media:.2f}")
plt.axvline(mediana, color='blue', linestyle='--', label=f"Mediana = {mediana:.2f}")
plt.axvline(moda, color='green', linestyle='--', label=f"Moda = {moda:.2f}")

plt.title("Distribución de Ingresos Simulados")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")

plt.legend()
plt.show()
```

### <span style="color:#2F749F;"><strong>Ejercicio 2: Cálculo de medidas de dispersión</strong></span>

A partir de la muestra generada en el Ejercicio 1, analiza qué tan dispersos están los datos respecto a su media. Calcula:
1. Rango
2. Varianza
3. Desviación estándar
4. Coeficiente de variación
5. Genera una gráfica con los resultados obtenidos (ver imagen de ejemplo)

<p align="center">
    <img src="https://github.com/mariabda2/intro_data_2026/blob/main/sesiones_practicas/imgs/sp_2_e2.png?raw=1" alt="sp_2_e2" width="900"  height="400">
</p>

- Rango: Mide la distancia entre el valor máximo y el mínimo para conocer qué tan amplio es el intervalo de los datos.

- Varianza: Mide cuánto se alejan los datos de la media en promedio.

- Desviación estándar: Se calcula mediante la raíz cuadrada de la varianza e indica cuánto se aleja típicamente un dato de la media.

- Coeficiente de variación: Mide la dispersión relativa respecto a la media.

```python
# Medidas de dispersión

rango = np.max(ingresos) - np.min(ingresos)
varianza = np.var(ingresos)
desv_est = np.std(ingresos)
coef_var = desv_est / media

print("Rango:", round(rango,2))
print("Varianza:", round(varianza,2))
print("Desviación estándar:", round(desv_est,2))
print("Coeficiente de variación:", round(coef_var,2))
```

```python
plt.figure(figsize=(10,5))

sns.histplot(ingresos, bins=10, kde=True, color='lightblue')

# Líneas importantes
plt.axvline(media, color='blue', linewidth=2, label='Media')
plt.axvline(media + desv_est, color='red', linestyle='--', label='+1 Desv. Est.')
plt.axvline(media - desv_est, color='red', linestyle='--', label='-1 Desv. Est.')
plt.axvline(np.min(ingresos), color='gray', linestyle='--', label='Mínimo')
plt.axvline(np.max(ingresos), color='gray', linestyle='--', label='Máximo')

# Caja de texto con resultados
texto = (
    f"Rango: {rango:.2f}\n"
    f"Varianza: {varianza:.2f}\n"
    f"Desv. estándar: {desv_est:.2f}\n"
    f"Coef. variación: {coef_var:.2f}"
)

plt.text(0.02, 0.95, texto, transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top',
         bbox=dict(boxstyle="round", facecolor="white", alpha=0.7))

plt.title("Distribución de ingresos simulados — Medidas de dispersión")
plt.xlabel("Ingreso mensual (miles de pesos)")
plt.ylabel("Frecuencia")

plt.legend()
plt.show()
```

### <span style="color:#2F749F;"><strong>Ejercicio 3: Cálculo de medidas de distribución</strong></span>

A partir de la muestra generada en el Ejercicio 1, analiza la distribución de los datos. Calcula:

2. Curtosis
3. Realiza la comparación con la distribución normal (ver la imagen de ejemplo)

- Asimetría: Mide si la distribución presenta una inclinación hacia algún lado.

- Curtosis: Mide la forma de la distribución comparada con la normal.

```python
# Medidas de distribución

asimetria = stats.skew(ingresos)
curtosis = stats.kurtosis(ingresos)

print("Asimetría:", round(asimetria,2))
print("Curtosis:", round(curtosis,2))
```

### <span style="color:#2F749F;"><strong>Ejercicio 5: Análisis de correlación</strong></span>

En esta segunda parte, genera **dos variables** a partir de tú número de cédula y realiza un análisis **bivariado** para calcular la relación entre ellas.

1. Use los **6 últimos dígitos** de su cédula como base (similar al Ejercicio 1).  
2. Genere dos variables aleatorias (ejemplo: `ingresos` y `gastos`) con distribución normal, usando esa base como **semilla reproducible**.  
3. Calcule los coeficientes de correlación de **Pearson** y **Spearman**.
4. Grafique las variables analizadas (ver la imagen de ejemplo)
5. Interprete los resultados.  

<p align="center">
    <img src="https://github.com/mariabda2/intro_data_2026/blob/main/sesiones_practicas/imgs/sp_2_e5.png?raw=1" alt="sp_2_e5" width="900"  height="400">

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

# Semilla (usa los últimos 6 dígitos de tu cédula)
np.random.seed(246638)

# Generar datos
n = 30
ingresos = np.random.normal(5, 2, n)
gastos = ingresos * 0.7 + np.random.normal(0, 1, n)

# Calcular correlaciones
pearson_corr, _ = pearsonr(ingresos, gastos)
spearman_corr, _ = spearmanr(ingresos, gastos)

# Gráfica
plt.figure(figsize=(8,5))
plt.scatter(ingresos, gastos)

plt.title("Relación entre Ingresos y Gastos")
plt.xlabel("Ingresos")
plt.ylabel("Gastos")

# Mostrar coeficientes en la gráfica
texto = f"Coef. Pearson = {pearson_corr:.3f}\nCoef. Spearman = {spearman_corr:.3f}"
plt.text(min(ingresos), max(gastos)-0.5, texto,
         bbox=dict(facecolor='lightgray', alpha=0.8))

plt.grid(True)
plt.show()
```

#INTERPRETACION
Existe una relacion fuerte entre Ingresos y Gastos. Esto significa que cuando los ingresos aumentan, los gastos tambien suelen aumentar

### <span style="color:#2F749F;"><strong>Ejercicio 6: Chi-Cuadrado</strong></span>

En esta parte, analiza si existe **asociación estadística** entre dos variables categóricas simuladas a partir de tu número de cédula.

1. Usa los **6 últimos dígitos** de tu cédula como base para generar una semilla reproducible.
2. Simula una muestra de **60 personas** con dos variables categóricas:
   - `Nivel educativo`: Básico, Medio, Superior.
   - `Acceso a internet`: Sí, No.
3. Crea una **tabla de contingencia** entre ambas variables.
4. Aplica la **prueba de Chi cuadrado de independencia**.
5. Visualiza la distribución conjunta con un mapa de calor (ver la imagen de con la información de ejemplo).
6. Interpreta el resultado.

<p align="center">
    <img src="https://github.com/mariabda2/intro_data_2026/blob/main/sesiones_practicas/imgs/sp_2_e6.png?raw=1" alt="sp_2_e6" width="900"  height="700">
</p>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

np.random.seed(246638)

n = 60

nivel_educativo = np.random.choice(["Básico","Medio","Superior"], n)
acceso_internet = np.random.choice(["No","Sí"], n)

df = pd.DataFrame({
    "Nivel educativo": nivel_educativo,
    "Acceso a internet": acceso_internet
})

# Tabla de contingencia
tabla = pd.crosstab(df["Nivel educativo"], df["Acceso a internet"])

print("Tabla de contingencia:")
print(tabla)

# Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(tabla)

# Frecuencias esperadas
expected_df = pd.DataFrame(expected, index=tabla.index, columns=tabla.columns)

# Heatmap con valores dentro de la tabla
plt.figure(figsize=(7,4))
sns.heatmap(
    tabla,
    annot=True,      # muestra números dentro de las celdas
    fmt="d",         # formato entero
    cmap="Blues",
    cbar=True,
    linewidths=0.5
)

plt.title("Distribución conjunta: Nivel educativo vs. Acceso a internet")
plt.xlabel("Acceso a internet")
plt.ylabel("Nivel educativo")

plt.show()

print("\nChi² =", round(chi2,2),
      ", p-valor =", round(p,3),
      ", grados de libertad =", dof)

print("\nFrecuencias esperadas:")
print(expected_df.round(1))
```

#INTERPRETACION
Con el p-valor nos damos cuenta que se rechaza la hipotesis de independencia, por tanto se tiene una relacion significativa entre el acceso a internet y nivel educativo
